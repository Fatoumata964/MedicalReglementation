```python
import time
from selenium import webdriver

with open("medecines_regulation.csv", "w") as file:
    file.write("title; inn_common; therap_area; date_authorisation; revision; last_update \n")
    
driver = webdriver.Chrome(executable_path = "C:\webdrivers\chromedriver.exe")

driver.get("https://www.ema.europa.eu/en/search?f%5B0%5D=ema_med_status%3A100108&f%5B1%5D=ema_med_status%3A100109&f%5B2%5D=ema_med_status%3A100111&f%5B3%5D=ema_medicine_bundle%3Aema_medicine&f%5B4%5D=ema_search_categories%3A83&f%5B5%5D=ema_search_categories%3A84")
driver.maximize_window()
time.sleep(2)

# Accept all cookies
l=driver.find_element("link text", "Accept all cookies")
l.click()
time.sleep(2)

# close button
b =driver.find_element('xpath', '//*[@id="cookie-consent-banner"]/div/div/div[2]/button')
b.click()

# drop withdrawn
w =driver.find_element('xpath', '/html/body/div[2]/div/main/div/div/div[3]/div[1]/div[2]/div/a[3]/span')
w.click()
time.sleep(1)

# drop refused
w =driver.find_element('xpath', '/html/body/div[2]/div/main/div/div/div[3]/div[1]/div[2]/div/a[3]/span/span[1]')
w.click()
time.sleep(2)

# drop veterinary
w =driver.find_element('xpath', '/html/body/div[2]/div/main/div/div/div[3]/div[1]/div[2]/div/a[5]/span')
w.click()

```

```python
urls = []
condition = True
while condition:
    listMed = driver.find_elements('xpath', "//div[contains(@class, 'teaser-title card-title bcl-heading')]")
    for el in listMed:
        url = el.find_element('css selector', 'a')
        urls.append(url.get_property('href'))
        
        try:
            driver.find_elements('xpath', '//*[@id="block-system-main-block"]/div/div/nav/ul/li[10]/a')[-1].text
        except:
            condition = False
            
```

```python
urls
```

```python
from tqdm import tqdm

descriptions = []
for i in urls:
    driver.get(i)
    name = driver.find_element('xpath', '//div[@class="heading-title"]/h1/span').text
    active_substance = driver.find_element('xpath', '//*[@id="block-system-main-block"]/article/div[1]/div[2]/div/div[3]/div[1]/dl/dd[2]').text
    common_name = driver.find_element('xpath', '//*[@id="block-system-main-block"]/article/div[1]/div[2]/div/div[3]/div[1]/dl/dd[3]').text
    therap_area = driver.find_element('xpath', '//*[@id="block-system-main-block"]/article/div[1]/div[2]/div/div[3]/div[1]/dl/dd[4]').text
    therapeutic_indication = driver.find_elements('xpath', '//*[@id="block-system-main-block"]/article/div[1]/div[2]/div/div[3]/div[3]')[0].text # + ' ' + driver.find_elements('xpath', '//*[@id="block-system-main-block"]/article/div[1]/div[2]/div/div[3]/div[3]')[1].text
    
    
    # Medicament used for
    us = driver.find_element('xpath', "//span[contains(text(),'used for?')]")
    driver.execute_script('arguments[0].scrollIntoView(true)', us)
    time.sleep(1)
    us.find_element('xpath', '../..').click()
    us.find_elements('xpath', '../../..')[0].text
    used_for = us.find_elements('xpath', '../../..')[0].text  #+ ' ' + p.find_elements('css selector', 'p')[1].text
    time.sleep(1)
    
    # Medicament usage
    usa = driver.find_element('xpath', "//span[contains(text(),'used?')]")
    driver.execute_script('arguments[0].scrollIntoView(true)', usa)
    time.sleep(1)
    usa.find_element('xpath', '../..').click()
    usa.find_elements('xpath', '../../..')[-1].text
    usage = usa.find_elements('xpath', '../../..')[-1].text
    time.sleep(1)
    

    # Medicament risks
    usb = driver.find_element('xpath', "//span[contains(text(),'risk')]")
    driver.execute_script('arguments[0].scrollIntoView(true)', usb)
    time.sleep(1)
    usb.find_element('xpath', '../..').click()
    usb.find_elements('xpath', '../../..')[0].text
    risk = usb.find_elements('xpath', '../../..')[0].text
    
    prodJ = {
        'name' : name,
        'active_substance' : active_substance,
        'common_name' : common_name,
        'therap_area' : therap_area,
        'therapeutic_indication' : therapeutic_indication,
        'used_for' : used_for,
        'usage': usage,
        'risk' : risk
    }
    descriptions.append(prodJ)
```
