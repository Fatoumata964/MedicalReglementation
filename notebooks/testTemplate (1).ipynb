```python
def generate_prompt(drug_name):
    """Generates a prompt for generating a drug information leaflet.

    Args:
        drug_name (str): Name of the drug.

    Returns:
        str: Generated prompt.
    """


    missing_parts = {
        "partie1": '''#Encadré. Cette partie doit conseiller au patient de lire attentivement la notice, et donner des conseils generaux en cas d'effets indesirables.''',
        "partie2": '''# Que contient cette notice? Cette partie doit refleter le plan de la notice.''',
        "partie3": f"# Qu'est ce que {drug_name} et comment l'utilisé?",
        "partie4": f"# Quels sont les informations a connaitre avant de prendre {drug_name}?",
        "partie5": f"# Comment prendre {drug_name} ?",
        "partie6": f"# Quels sont les effets indesirables eventuelles de {drug_name}?",
        "partie7": f"# Comment conserver {drug_name}?",
        "partie8": '''# Contenu de l'emballage et autres informations. Cette partie doit citer les substances actives et les autres composants sur l'emballage. Les autres informations porteront sur le titulaire, l'exploitant de l'autorisation de mise sur le marché et le fabricant,...'''
    }

    text = ""


    for part_name, part_content in missing_parts.items():
        if part_name not in text:
            text += f"\n**Attention:** N'oubliez pas d'inclure la {part_name} :\n{part_content}\n"


    prompt_template = f"""
    ## Notice d'information patient - {drug_name}

    {text}

    **Objectif:** Générer une notice d'information patient complète et informative pour le médicament {drug_name}.

    **Instructions:** Suivez les directives et les conseils mentionnés dans les rappels ci-dessus pour inclure toutes les sections nécessaires dans la notice.
     """
    prompt = PromptTemplate(input_variables=["drug_name"], template=prompt_template)
    return prompt
```
