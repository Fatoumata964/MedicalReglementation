```python
fulldf1 = fulldf.copy()
fulldf1['Condition / indication'] = fulldf1['Condition / indication'].astype('str')
translator = GoogleTranslator(source='auto', target='fr')

# translate columns' name using rename function
fulldf1.rename(columns=lambda x: translator.translate(x), inplace=True)

# translated column names
fulldf1.columns
```

```python
import nltk

translations = {}
col = (fulldf1.applymap(type) == str).all(0)
d1f = fulldf1.loc[: , col]
for column in d1f.columns:
    # unique elements of the column
    unique_elements = d1f[column].unique()
    for element in unique_elements:
        if len(element)<5000:
            # add translation to the dictionary
            translations[element] = translator.translate(element)

        else:
            x = nltk.tokenize.sent_tokenize(element)
            sentences = []
            for sentence in x:
                sentence = translator.translate(sentence)
                sentences.append(sentence)
                translations[element] = sentences

print(translations)
```
